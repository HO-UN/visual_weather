```python
import pandas as pd
import pyecharts.options as opts
from pyecharts.charts import WordCloud
data = pd.read_csv('北上广深历史天气.csv')
data
```

```python
def cloudtu(cs):
    freq = {}
    d1=data[data['城市']==cs].天气.values.tolist()
    for i in d1:
        if("~"in i):
            if(i.split("~")[0] in freq):
                freq[i.split("~")[0]]+=1
            else:
                freq[i.split("~")[0]]=1
            if(i.split("~")[1] in freq):
                freq[i.split("~")[1]]+=1
            else:
                freq[i.split("~")[1]]=1
        else:
            if(i in freq):
                freq[i]+=1
            else:
                freq[i]=1
    c = (
        WordCloud()
        .add(
            series_name="{}天气云图".format(cs),
            # 系列数据项，[(word1, count1), (word2, count2)]
            data_pair=list(freq.items()),
            # 单词字体大小范围
            word_size_range=[20, 100])
        # 全局配置项
        .set_global_opts(
            # 标题设置
            title_opts=opts.TitleOpts(
                title="{}天气云图".format(cs), title_textstyle_opts=opts.TextStyleOpts(font_size=23)
            ),
            # 提示框设置
            tooltip_opts=opts.TooltipOpts(is_show=True),
        )
        .render_notebook()
    )
    return c

  
       
```

```python
cloudtu('广州')
```

```python
cloudtu('北京')
```

```python
cloudtu('上海')
```
